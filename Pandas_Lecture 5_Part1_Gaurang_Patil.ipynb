### How to Iterate Over Rows

#### Exercise 1 Iterating DataFrames with items( )

```python
import pandas as pd
import warnings 
  
# Settings the warnings to be ignored 
warnings.filterwarnings('ignore') 

dataframe = pd.DataFrame({
    'first_name' : ['John', 'Jane', 'Marry', 'Victoria', 'Gabriel', 'Layla'],
    'last_name' : ['Smith', 'Doe', 'Jackson', 'Smith', 'Brown', 'Martinez'],
    'age' : [34, 29, 37, 52, 26, 32]},
    index = ['id001', 'id002', 'id003', 'id004', 'id005', 'id006'])

print(dataframe)
```

```python
# --- Added the code here ---
for col_name, data in dataframe.items():
    print("col_name", col_name, "\ndata:", data)
# ---------------------------
```

```python
# particular row as per requirement
for col_name, data in dataframe.items():
    # --- Added the code here ---
    print("col_name: ", col_name, "\ndata: ", data[1])
    # ---------------------------
```

```python
for col_name, data in dataframe.items():
    # --- Added the code here ---
    print("col_name: ", col_name, "\ndata: ", data['id002'])
    # ---------------------------
```

#### Exercise 2 Iterating DataFrames with iterrows( )

```python
import pandas as pd

dataframe = pd.DataFrame({
    'first_name' : ['John', 'Jane', 'Marry', 'Victoria', 'Gabriel', 'Layla'],
    'last_name' : ['Smith', 'Doe', 'Jackson', 'Smith', 'Brown', 'Martinez'],
    'age' : [34, 29, 37, 52, 26, 32]},
    index = ['id001', 'id002', 'id003', 'id004', 'id005', 'id006'])

print(dataframe)
```

```python
# --- Added the code here ---
for i, row in dataframe.iterrows():
    print(f"Index: (i)")
    print(f"{row}\n")
# ---------------------------
```

```python
for i, row in dataframe.iterrows():
    print(f"Index: {1}")
    # --- Added the code here ---
    print(f"{row[0]}")
    # ---------------------------
```

```python
for i, row in dataframe.iterrows():
    print(f"Index: {i}")
    # --- Added the code here ---
    print(f"{row['first_name']}")
    # ---------------------------
```

#### Exercise 3 Iterating DataFrames with itertuples( )

```python
import pandas as pd

dataframe = pd.DataFrame({
    'first_name' : ['John', 'Jane', 'Marry', 'Victoria', 'Gabriel', 'Layla'],
    'last_name' : ['Smith', 'Doe', 'Jackson', 'Smith', 'Brown', 'Martinez'],
    'age' : [34, 29, 37, 52, 26, 32]},
    index = ['id001', 'id002', 'id003', 'id004', 'id005', 'id006'])

print(dataframe)
```

#### Exercise 9 Merge DataFrames Using concat( )

```python
import pandas as pd

df1 = pd.DataFrame(
    {'user_id' : ['id001', 'id002', 'id003', 'id004', 'id005', 'id006', 'id007'],
     'first_name' : ['Rivi', 'Wynnie', 'Kristos', 'Madalyn', 'Tobe', 'Regan', 'Kristin'],
     'last_name' : ['Valti', 'McMurty', 'Ivabets', 'Max', 'Riddich', 'Huyghe', 'Illis'],
     'email' : ['rvalti0@example.com',
                'wmcmurty1@example.com',
                'kivanets2@example.com',
                'mmax3@example.com',
                'triddich4@example.com',
                'rhuyghe@example.com',
                'killis4@example.com'] 
    }
)

df2 = pd.DataFrame({
    'user_id' : ['id001', 'id002', 'id003', 'id004', 'id005'],
    'image_url' : ['http://example.com/img/id001.png',
                   'http://example.com/img/id002.jpg',
                   'http://example.com/img/id003.bmp',
                   'http://example.com/img/id004.jpg',
                   'http://example.com/img/id005.png',]
})
```

```python
df2_addition = pd.DataFrame({
    'user_id' : ['id006', 'id007'],
    'image_url' : ['http://example.com/img/id006.png',
                   'http://example.com/img/id007.jpg']
})
```

```python
df2_addition
```

```python
# --- Added the code here ---
df_row_concat = pd.concat([df2, df2_addition])
# ---------------------------
```

```python
df_row_concat
```

```python
# --- Added the code here ---
df_row_concat = pd.concat([df2, df2_addition], ignore_index = True)
# ---------------------------
```

```python
df_row_concat
```

```python
# --- Added the code here ---
df_column_concat = pd.concat([df1, df_row_concat], axis = 1)
# ---------------------------
```
